#### 2.2 Create a correlation heatmap using seaborn:

```python
# Create a subset to exclude unnessary columns
sub = df_drop.drop(columns=['Happiness Rank', 'Year'])
```

```python
sub
```

```python
plt.close('all')
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
plt.show()
```

```python
plt.close()
```

Happiness Score Correlations:

1. Strong positive correlation with GDP per capita (0.79): Wealthier countries tend to have higher happiness scores.

2. Moderate positive correlation with Life Expectancy (0.74) and Social Support (0.65): Longer life expectancy and stronger social networks are linked to higher happiness.

3. Moderate correlation with Freedom (0.55): Freedom to make life choices also contributes to happiness, though less strongly than GDP or Life Expectancy.

4. Weaker correlation with Trust (0.4) and Generosity (0.14): These factors still contribute, but their impact on happiness is less significant compared to others.

5. Dystopia Residual (0.47): Represents unexplained components contributing to happiness, moderately correlated.

```python
# Export the heat map to the folder
corr.figure.savefig(os.path.join(path, '04 Visualizations', 'correlation_heatmap.png'), dpi=300, bbox_inches='tight')
```

#### 2.3 Scatterplot

```python
# Create a scatterplot for the "Happiness Score" and "GDP per capita" columns in seaborn

scatter = sns.lmplot(x = 'GDP per capita', y = 'Happiness Score', data = df)
plt.show()
```

Insights:
1. The scatterplot and regression line indicate a clear strong positive correlation between "Happiness Score" and "GDP per capita". This means countries with higher GDP per capita generally tend to have higher happiness scores.
 
2. Clustering of Points:
Most of the points are clustered along the line, indicating a strong linear relationship between GDP per capita and Happiness Score. However, a few outliers have low GDP but high Happiness Scores, and a few outliers have high GDP but relatively low Happiness Score. 

3. Variance:
While the general trend is positive, there is some spread of happiness scores for countries with similar GDP per capita, indicating that other factors also influence happiness levels.

```python
# Export the heat map to the folder
scatter.figure.savefig(os.path.join(path, '04 Visualizations', 'scatterpl_gdp_score.png'), dpi=300, bbox_inches='tight')
```

```python
plt.close()
```

```python
# Create a scatterplot for the "Happiness Score" and "Generosity" columns in seaborn

scatter = sns.lmplot(x = 'Generosity', y = 'Happiness Score', data = df)
plt.show()
```

Insights
1. Weak Positive Correlation: 
There is a slight positive trend, but the relationship between Generosity and Happiness Score is not very strong. This is evident from the loosely scattered points around the line.

2. Clustered Values: Most data points are clustered around a Generosity value of 0 to 0.2, indicating that many countries have similar levels of generosity.

3. Outliers: There are a few outliers with higher Generosity values (above 0.6), but these do not seem to significantly affect the overall trend.

4. Impact on Happiness: Generosity appears to have a minimal impact on Happiness Score compared to other variables, as indicated by the weak slope of the line.

```python
# Export the scatterplot to the folder
scatter.figure.savefig(os.path.join(path, '04 Visualizations', 'scatterpl_generosity_score.png'), dpi=300, bbox_inches='tight')
```

#### 2.5 Category Plots

##### Creating a categorical variable that splits the "Happiness Score" column into categories

```python
# Use a histogram to visualize the distribution of the variables.
cate = sns.histplot(df['Happiness Score'], bins = 20 , kde = True)
plt.show()
```

```python
# Export the category plot to the folder
cate.figure.savefig(os.path.join(path, '04 Visualizations', 'category_score.png'), dpi=300, bbox_inches='tight')
```

```python
plt.close()
```

```python
df.loc[df['Happiness Score'] < 4, 'Score category'] = 'Low score'
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] < 6), 'Score category'] = 'Middle score'
df.loc[df['Happiness Score'] >= 6, 'Score category'] = 'High score'
```

```python
# Check the counts of score categories
df['Score category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the score categories created above

sns.set(style="ticks")
cate_2 = sns.catplot(x="Life Expectancy", y="Happiness Score", hue="Score category", data=df)
plt.show()
```

The categories (green, blue, orange) show a clearer separation, with each happiness category occupying distinct regions of the plot. This reflects the stronger correlation between happiness and life expectancy.

```python
# Export the category plot
cate_2.figure.savefig(os.path.join(path, '04 Visualizations', 'category_life_score.png'), dpi=300, bbox_inches='tight')
plt.close()
```

```python
# Create another categorical plot in seaborn using the score categories created above

sns.set(style="ticks")
cate_3 = sns.catplot(x="Freedom", y="Happiness Score", hue="Score category", data=df)
plt.show()
```

The categories overlap more significantly, particularly between middle and high scores. This suggests that while freedom impacts happiness, its effect is less pronounced compared to life expectancy.

```python
# Export the category plot
cate_3.figure.savefig(os.path.join(path, '04 Visualizations', 'category_freedom_score.png'), dpi=300, bbox_inches='tight')
plt.close()
```
